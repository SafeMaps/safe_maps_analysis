```python
# import gmplot package 
import gmplot 


#Reading lat.txt and long.txt into arrays

lat_file = open("../lat.txt", "r")
long_file = open("../long.txt", "r")
latitude_list = list(map(float,lat_file.read().split(',')))
longitude_list = list(map(float,long_file.read().split(',')))


# GoogleMapPlotter return Map object 
# Pass the center latitude and 
# center longitude 
gmap1 = gmplot.GoogleMapPlotter(40.760292, 
                                -73.996129, 13 ) 


currentL = [40.740292,-73.996129]
destinationL = [40.763242,-73.976942]


#Display current location
gmap1.scatter([currentL[0]],[currentL[1]], '#00fff9', 
                              size = 30, marker = False ) 

#Display destination location
gmap1.scatter([destinationL[0]],[destinationL[1]], '#00ff00', 
                              size = 30, marker = False ) 

# Draw Box
gmap1.plot([currentL[0],destinationL[0]],[currentL[1],currentL[1]],'cornflowerblue', edge_width = 2.5) 
gmap1.plot([currentL[0],currentL[0]],[currentL[1],destinationL[1]],'cornflowerblue', edge_width = 2.5) 
gmap1.plot([destinationL[0],destinationL[0]],[destinationL[1],currentL[1]],'cornflowerblue', edge_width = 2.5) 
gmap1.plot([destinationL[0],currentL[0]],[destinationL[1],destinationL[1]],'cornflowerblue', edge_width = 2.5) 


# display the crimes to the map
gmap1.scatter( latitude_list, longitude_list, '#ff0000', 
                              size = 10, marker = False ) 

# Pass the absolute path 
gmap1.draw("./file.html" ) 
```

```python
import sys, json, gmplot

test = '[{"latitude":40.754877681,"longitude":-73.995814518},{"latitude":40.741125534,"longitude":-73.984588781},{"latitude":40.741341373,"longitude":-73.978392609}]'
crimes = json.loads(test)
latitude_list = []
longitude_list = []

for crime in crimes:
    latitude_list.append(crime['latitude'])
    longitude_list.append(crime['longitude'])
    
# GoogleMapPlotter return Map object 
# Pass the center latitude and 
# center longitude 
gmap1 = gmplot.GoogleMapPlotter(40.760292, 
                                -73.996129, 13 ) 


currentL = [40.740292,-73.996129]
destinationL = [40.763242,-73.976942]


#Display current location
gmap1.scatter([currentL[0]],[currentL[1]], '#00fff9', 
                              size = 30, marker = False ) 

#Display destination location
gmap1.scatter([destinationL[0]],[destinationL[1]], '#00ff00', 
                              size = 30, marker = False ) 

# Draw Box
gmap1.plot([currentL[0],destinationL[0]],[currentL[1],currentL[1]],'cornflowerblue', edge_width = 2.5) 
gmap1.plot([currentL[0],currentL[0]],[currentL[1],destinationL[1]],'cornflowerblue', edge_width = 2.5) 
gmap1.plot([destinationL[0],destinationL[0]],[destinationL[1],currentL[1]],'cornflowerblue', edge_width = 2.5) 
gmap1.plot([destinationL[0],currentL[0]],[destinationL[1],destinationL[1]],'cornflowerblue', edge_width = 2.5) 


# display the crimes to the map
gmap1.scatter( latitude_list, longitude_list, '#ff0000', 
                              size = 10, marker = False ) 

# Pass the absolute path 
gmap1.draw("./vizualization/file.html" ) 
```
